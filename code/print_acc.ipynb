```python
import math

def is_composite(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return True
    return False

number = 9293

if is_composite(number):
    print("Yes, 9293 is a composite number.")
else:
    print("No, 9293 is not a composite number.")
```
