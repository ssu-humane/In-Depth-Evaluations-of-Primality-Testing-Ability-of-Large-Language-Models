```python
import random
import json 
from utils import chatgpt, write_output 
random.seed(1024)

with open("../json/2_digit_prime.json") as file:
    digit_2_primality = json.load(file)
try:
    digit_2_primality = random.sample(digit_2_primality, 500)
except:
    pass 

max_len = len(digit_2_primality)
f1 = open("../output/2_digit_output.txt", "r", encoding='UTF-8')
f2 = open("../yes_no/digit_2_primality_output.txt", "r", encoding='UTF-8')
text2 = f2.read()
text1 = f1.read()
start_idx = text1.find("###")

for i in range(text2.count("###")):
    end_idx   = text1.find("###", start_idx+1)
    output = text1[start_idx:end_idx]
    start_idx = end_idx

while(1):
    try:
        f2 = open("../yes_no/digit_2_primality_output.txt", "r", encoding='UTF-8')
        text2 = f2.read()
        x_idx = text2.count("###") // 4
        q_idx = text2.count("###") %  4

        for x in digit_2_primality[x_idx:]:
            for i in range(q_idx, 4):
                end_idx   = text1.find("###", start_idx+1)
                output = text1[start_idx:end_idx]
                question = (output + x[f'question{i+1}'].replace("Let's think step by step.", "") + "\nAnswer with either Yes or No.").replace("###", " ")
                write_output(question, '../yes_no/digit_2_primality_output.txt')
                q_idx = 0
                start_idx = end_idx
                
        f2 = open("../yes_no/digit_2_primality_output.txt", "r", encoding='UTF-8')
        text2 = f2.read()           
        if text2.count("###") == 4 * max_len:
            break
    except:
            pass 
```
